# Phase 1: Data Preparation & Exploratory Analysis
**Objective:** Load the superstore dataset, understand its structure, check for data quality issues, and perform initial transformations.

```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

```python
# Load the CSV file
file_path = '../data/raw/superstore_raw.csv'
df = pd.read_csv(file_path)
```

```python
# 1. Show basic information
print(f"Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")
print("\n--- Data Types & Columns ---")
print(df.info())

# Display the first 10 rows
display(df.head(10))

# Display the last 10 rows
display(df.tail(10))
```

```python
# 2. Check for missing values and duplicates
print("\n--- Missing Values Check ---")
missing_values = df.isnull().sum()
if missing_values.sum() > 0:
    print(missing_values[missing_values > 0])
else:
    print("No missing values found in the dataset.")

duplicates_count = df.duplicated().sum()
print(f"\nTotal duplicate rows found: {duplicates_count}")
```

```python
# 3. Summary statistics for numerical columns
print("\n--- Summary Statistics (Numerical) ---")
display(df.describe())

# Summary statistics for categorical columns
print("\n--- Summary Statistics (Categorical) ---")
display(df.describe(include=['O', 'category']))
```

```python
# 4. Check unique values for important categorical columns
important_categorical_cols = ['Segment', 'Category', 'Region', 'Ship Mode']

print("\n--- Unique Values & Frequencies ---")
for col in important_categorical_cols:
    print(f"\n{col} ({df[col].nunique()} unique segments):")
    print(df[col].value_counts())
    print("-" * 30)
```

```python
# 5. Convert 'Order Date' and 'Ship Date' to datetime format
# The dataset dates are typically formatted as MM/DD/YYYY
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

print("\n--- Date Columns Conversion Check ---")
print(df[['Order Date', 'Ship Date']].dtypes)
```

```python
# 6. Create initial time-based columns from 'Order Date'
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
# Provide Quarter formatting like Q1, Q2 etc.
df['Order Quarter'] = 'Q' + df['Order Date'].dt.quarter.astype(str)

print("\n--- New Time-based Columns ---")
display(df[['Order Date', 'Order Year', 'Order Month', 'Order Quarter']].head())
```
